```python
import cv2
import numpy as np

# Load the image
image = cv2.imread('floorPlan_0.png', cv2.IMREAD_GRAYSCALE)

# Apply the Sobel operator for edge detection
vertical_edges = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
horizontal_edges = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)

# Calculate the edge magnitude
edge_magnitude = np.sqrt(vertical_edges**2 + horizontal_edges**2)

# Apply thresholding to identify edges
threshold_value = 200  # You can adjust this value
edge_image = (edge_magnitude > threshold_value).astype(np.uint8) * 255

# Display the resulting edge image
cv2.imshow('Edge Image', edge_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
